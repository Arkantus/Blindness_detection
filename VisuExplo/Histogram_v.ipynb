```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from PIL import Image
from pathlib import Path
```

```python
imagesPath = Path("../train_images")
imgesIte = imagesPath.iterdir()
```

```python
singleImageHistogram = []
fp = open("histos.bk",'w')
for i in imgesIte:
    if str(i).endswith(".png"):
        img = Image.open(i)
        sih = img.histogram()
        singleImageHistogram.append(sih)
        fp.write(str(sih))
fp.close()
```

```python
array = np.array(singleImageHistogram)
import seaborn as sns; sns.set()
ax = sns.heatmap(array)
```

```python
means = []
stdds = []
for h in singleImageHistogram:
    means.append(np.mean(h))
    stdds.append(np.std(h))
```

```python
sns.lineplot(y = means, x = list(range(len(means))))
```
